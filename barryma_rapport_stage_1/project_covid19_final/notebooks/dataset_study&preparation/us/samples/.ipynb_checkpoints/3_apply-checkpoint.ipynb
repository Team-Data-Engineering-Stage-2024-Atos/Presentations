## Let's focus on the first row (cell too)

```python
import pandas as pd
import csv
import re

df = pd.read_csv('RAW_us_deaths_sample.csv')


print(df["Combined_Key"])
```

```python
df.drop('Combined_Key', axis=1, inplace=True)
df.head()
```

```python
input_string = df["Province_State"][0]
print(input_string)
```

```python
parts = input_string.split('\"')

my_dict = {}

for i, part in enumerate(parts, 1):
    # Split the part by comma, remove double quotes and filter out empty strings
    list_part = [elm.replace('\"', '') for elm in part.split(',') if elm]
    
    # If the list has less than 3 elements, add it to the dictionary
    if len(list_part) != 3:
        my_dict[str(i)] = list_part

# Flatten the dictionary values into a list
values_list = [val for sublist in my_dict.values() for val in sublist]

columns = (df.columns).tolist()
values = values_list


print("Columns : \n", columns)
print("***********  ---------   *************"*10)
print("Values : \n", values)
```

```python
print("Columns length : \n", len(columns))
print("***********  ---------   *************"*10)
print("Values length : \n", len(values))
```

### Let's map these two lists with columns as key and values as value

```python
col_value_list = list(zip(columns, values))

print(col_value_list)
```

## Let's apply this solution with the remaining rows of the dataset

```python
import pandas as pd
import csv
import re

dataset_path = 'RAW_us_confirmed_cases_sample.csv'

DF = pd.read_csv(dataset_path)
DF.drop('Combined_Key', axis=1, inplace=True)
```

```python
DF.columns[:12] #Verify that Combined_key is dropped
```

```python
DF.shape[0]
```

```python
data = DF["Province_State"]
col_index_max = DF.shape[0]
columns = (DF.columns).tolist()
row_values = []
my_dict = {}

print("Columns : \n", columns)
print("***********  ---------   *************"*10)
    
    
for i in range(0, col_index_max):
    input_string = data[i]
    
    parts = input_string.split('\"')
    dict = {}

    for j, part in enumerate(parts, 1):
        # Split the part by comma, remove double quotes and filter out empty strings
        list_part = [elm.replace('\"', '') for elm in part.split(',') if elm]

        # If the list has less than 3 elements, add it to the dictionary
        if len(list_part) != 3:
            my_dict[str(j)] = list_part

    # Flatten the dictionary values into a list
    values_list = [val for sublist in my_dict.values() for val in sublist]

    row_values.append(tuple(values_list))

    
#print(row_values)

cleaned_us_list = [(key, *value) for key, value in zip(columns, row_values)]

print(cleaned_us_list)
```

## Let's end by creating a new clean csv

```python
new_dataset_path = 'cleaned_us_cases_sample.csv'

with open(new_dataset_path, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(DF.columns)
    writer.writerows(row_values)
```
